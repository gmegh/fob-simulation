# Generate dataset example

This notebook shows how to run the generate_dataset package in a notebook.

Import relevant python packages:

```python
import matplotlib.pyplot as plt
import generate_dataset as g
```

```python
durations = [0.015]
duration_label = ['15ms']
num_simulations = 15
intensities = [int(7.5e5)]

for idx in  range(len(durations)):

    for idy in range(len(intensities)):

        g.main(durations[idx], duration_label[idx], intensities[idy], num_simulations, atm = 'full', offset = 0)
```
