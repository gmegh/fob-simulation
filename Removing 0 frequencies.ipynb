```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack, ndimage
import math
from tqdm.notebook import tqdm
import seaborn as sns
import galsim
```

```python
name = 'FRB'
method = 'FFT'
duration = '150s'
num_phot = 750000
it = 0
```

```python
image = np.loadtxt('data/{}s/{}_{}_{}_{}_{}.txt'.format(name, name, method, duration, num_phot, it), delimiter = ',')
image[image < 0] = 0
img = image

rng = galsim.BaseDeviate()
gd = galsim.GaussianDeviate(rng, sigma = np.sqrt(1_000))
gd.add_generate(image)

plt.set_cmap('viridis')
plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
plt.title('Original image', fontsize = 20)
plt.imshow(image)
cbar = plt.colorbar(fraction = 0.046, pad = 0.04)
cbar.ax.tick_params(labelsize = 15)
plt.tick_params(labelsize = 15)

fft2 = np.fft.fftshift(np.fft.fft2(image))
ff = np.abs(fft2)
plt.subplot(1,2,2)
plt.title('2D FFT', fontsize = 20)
plt.imshow(ff)
cbar = plt.colorbar(fraction = 0.046, pad = 0.04)
cbar.ax.tick_params(labelsize = 15)
plt.tick_params(labelsize = 15)
```

```python
print(np.where(ff == np.max(ff)))
ff_rem = fft2
ff_rem[8:13, 8:13] = 0

plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
plt.title('2D FFT', fontsize = 20)
plt.imshow(abs(ff_rem))
cbar = plt.colorbar(fraction = 0.046, pad = 0.04)
cbar.ax.tick_params(labelsize = 15)
plt.tick_params(labelsize = 15)

image_rem = np.fft.ifft2(ff_rem)

plt.subplot(1,2,2)
plt.title('Reconstructed image', fontsize = 20)
plt.imshow(abs(image_rem))
cbar = plt.colorbar(fraction = 0.046, pad = 0.04)
cbar.ax.tick_params(labelsize = 15)
plt.tick_params(labelsize = 15)
```
